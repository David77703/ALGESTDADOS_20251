```python
#1 - Crie um algoritmo que simule um sistema de reservas de um restaurante. O programa deve solicitar o número de pessoas para a reserva e verificar se há disponibilidade, considerando que o restaurante possui apenas mesas de 4 lugares. Se o número de pessoas exceder a capacidade de uma mesa, o programa deve calcular quantas mesas serão necessárias e informar ao usuário.

def fazer_reserva():

  capacidade_mesa = 4
  mesas_disponiveis = 10

  try:
    num_pessoas = int(input("Quantas pessoas para a reserva? "))
  except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
    return

  if num_pessoas <= 0:
    print("O número de pessoas deve ser maior que zero.")
    return

  if num_pessoas <= capacidade_mesa:
    if mesas_disponiveis > 0:
      print(f"Reserva para {num_pessoas} pessoa(s) confirmada em uma mesa.")
    else:
      print("Desculpe, não há mesas disponíveis no momento.")
  else:
    mesas_necessarias = (num_pessoas + capacidade_mesa - 1) // capacidade_mesa
    if mesas_necessarias <= mesas_disponiveis:
      print(f"Para {num_pessoas} pessoas, serão necessárias {mesas_necessarias} mesa(s).")
      print("Reserva confirmada.")
    else:
      print("Desculpe, não há mesas suficientes para acomodar este grupo.")

fazer_reserva()
```

```python
#2 - Desenvolva um algoritmo que auxilie na gestão de estoque de uma loja. O programa deve permitir que o usuário insira o nome e a quantidade de diferentes produtos. Se a quantidade de um produto for inferior a 5 unidades, o sistema deve alertar que o estoque está baixo e sugerir a reposição. O processo deve continuar até que o usuário decida encerrar.

estoque = {}

def gerenciar_estoque():
  while True:
    nome_produto = input("Digite o nome do produto ou 'sair' para encerrar: ")
    if nome_produto.lower() == 'sair':
      break

    try:
      quantidade = int(input(f"Digite a quantidade de {nome_produto}: "))
      estoque[nome_produto] = quantidade

      if quantidade < 5:
        print(f"ATENÇÃO: O estoque de {nome_produto} está baixo ({quantidade} unidades). Sugerimos a reposição.")

    except ValueError:
      print("Entrada inválida para a quantidade. Por favor, digite um número inteiro.")

    print("Estoque atual:")
    for produto, qtd in estoque.items():
      print(f"{produto}: {qtd}")
    print("-" * 20)

gerenciar_estoque()
```

```python
#3 - Elabore um algoritmo que simule um quiz de perguntas e respostas. O programa deve apresentar uma série de perguntas ao usuário, oferecer opções de resposta e verificar se a resposta fornecida está correta. Ao final, o sistema deve exibir a pontuação total do usuário e fornecer um feedback baseado no desempenho.

def quiz():
  perguntas = [
      {
          "pergunta": "Qual a capital do Brasil?",
          "opcoes": ["A) Rio de Janeiro", "B) São Paulo", "C) Brasília", "D) Belo Horizonte"],
          "resposta_correta": "C"
      },
      {
          "pergunta": "Quanto é 7 * 8?",
          "opcoes": ["A) 56", "B) 49", "C) 64", "D) 72"],
          "resposta_correta": "A"
      },
      {
          "pergunta": "Quem pintou a Monalisa?",
          "opcoes": ["A) Vincent van Gogh", "B) Leonardo da Vinci", "C) Pablo Picasso", "D) Michelangelo"],
          "resposta_correta": "B"
      }
  ]

  pontuacao = 0

  print("Bem-vindo ao Quiz!")
  print("-" * 20)

  for item in perguntas:
    print(item["pergunta"])
    for opcao in item["opcoes"]:
      print(opcao)

    resposta_usuario = input("Digite a letra da sua resposta: ").strip().upper()

    if resposta_usuario == item["resposta_correta"]:
      print("Resposta correta!")
      pontuacao += 1
    else:
      print(f"Resposta incorreta. A resposta correta era: {item['resposta_correta']}")

    print("-" * 20)

  print("Quiz finalizado!")
  print(f"Sua pontuação total: {pontuacao}/{len(perguntas)}")

  if pontuacao == len(perguntas):
    print("Excelente! Você acertou todas as perguntas!")
  elif pontuacao >= len(perguntas) / 2:
    print("Bom trabalho! Você teve um bom desempenho.")
  else:
    print("Continue estudando! Você pode melhorar.")

quiz()
```

```python
#4 - Crie um algoritmo que simule um sistema de controle de acesso a um edifício. O programa deve solicitar ao usuário um código de acesso e verificar se ele é válido. Caso contrário, deve permitir novas tentativas até que o número máximo de 3 tentativas seja atingido. Se o acesso for concedido, uma mensagem de boas-vindas deve ser exibida; caso contrário, uma mensagem de bloqueio deve ser apresentada.

def controle_acesso():
  codigo_acesso_valido = "007113"
  tentativas_maximas = 3
  tentativas = 0

  while tentativas < tentativas_maximas:
    codigo_digitado = input("Digite o código de acesso: ")

    if codigo_digitado == codigo_acesso_valido:
      print("Acesso concedido. Bem-vindo!")
      return
    else:
      tentativas += 1
      tentativas_restantes = tentativas_maximas - tentativas
      if tentativas_restantes > 0:
        print(f"Código incorreto. Você tem mais {tentativas_restantes} tentativa(s).")

  print("Número máximo de tentativas atingido. Acesso bloqueado.")

controle_acesso()
```

```python
#5 - Desenvolva um algoritmo que ajude no planejamento de uma viagem de carro. O programa deve solicitar a distância total a ser percorrida, a capacidade do tanque de combustível do veículo e o consumo médio de combustível (km/l). Com base nessas informações, o sistema deve calcular quantas paradas para reabastecimento serão necessárias, assumindo que o motorista sempre reabastece o tanque completamente.

def planejar_viagem():
  try:
    distancia_total = float(input("Digite a distância total a ser percorrida (em km): "))
    capacidade_tanque = float(input("Digite a capacidade do tanque de combustível (em litros): "))
    consumo_medio = float(input("Digite o consumo médio de combustível (em km/l): "))
  except ValueError:
    print("Entrada inválida. Por favor, digite números.")
    return

  if distancia_total <= 0 or capacidade_tanque <= 0 or consumo_medio <= 0:
    print("Os valores digitados devem ser maiores que zero.")
    return

  autonomia_tanque = capacidade_tanque * consumo_medio

  if distancia_total <= autonomia_tanque:
    print("Não serão necessárias paradas para reabastecimento.")
  else:
    paradas_necessarias = int(distancia_total / autonomia_tanque)
    if distancia_total % autonomia_tanque == 0:
        paradas_necessarias -= 1
    print(f"Serão necessárias {paradas_necessarias} parada(s) para reabastecimento.")

planejar_viagem()
```

```python
#6 - Elabore um algoritmo que simule um sistema de avaliação de filmes. O programa deve permitir que o usuário insira o título de vários filmes e uma nota de 0 a 10 para cada um. Após o usuário finalizar as entradas, o sistema deve exibir a média das notas atribuídas e listar os filmes que receberam nota superior a 8, destacando-os como altamente recomendados.

def avaliar_filmes():
  filmes = {}
  soma_notas = 0

  print("Sistema de Avaliação de Filmes")
  print("Digite o título do filme e a nota (0 a 10). Digite 'sair' para finalizar.")
  print("-" * 30)

  while True:
    titulo_filme = input("Título do filme: ")
    if titulo_filme.lower() == 'sair':
      break

    try:
      nota = float(input(f"Nota para '{titulo_filme}' (0-10): "))
      if 0 <= nota <= 10:
        filmes[titulo_filme] = nota
        soma_notas += nota
      else:
        print("Nota inválida. Por favor, insira uma nota entre 0 e 10.")
    except ValueError:
      print("Entrada inválida. Por favor, digite um número para a nota.")
    print("-" * 30)

  if not filmes:
    print("Nenhum filme avaliado.")
    return

  media_notas = soma_notas / len(filmes)
  print(f"Média das notas atribuídas: {media_notas:.2f}")

  print("Filmes altamente recomendados (nota superior a 8):")
  altamente_recomendados = False
  for titulo, nota in filmes.items():
    if nota > 8:
      print(f"- {titulo} (Nota: {nota})")
      altamente_recomendados = True

  if not altamente_recomendados:
    print("Nenhum filme recebeu nota superior a 8.")

avaliar_filmes()
```

```python
#7 - Crie um algoritmo que simule um sistema de monitoramento de temperatura ambiente. O programa deve solicitar ao usuário que insira a temperatura registrada a cada hora durante um período de 24 horas. Ao final, o sistema deve calcular a temperatura média do dia, identificar a temperatura mais alta e a mais baixa registradas, e alertar se em algum momento a temperatura ultrapassou 30°C ou ficou abaixo de 15°C.

def monitorar_temperatura():
  temperaturas = []
  temperatura_minima_alerta = 15
  temperatura_maxima_alerta = 35

  print("Sistema de Monitoramento de Temperatura Ambiente")
  print("Por favor, insira a temperatura registrada a cada hora (24 horas).")
  print("-" * 40)

  for hora in range(24):
    while True:
      try:
        temp_hora = float(input(f"Temperatura na hora {hora + 1}: "))
        temperaturas.append(temp_hora)
        break
      except ValueError:
        print("Entrada inválida. Por favor, digite um número para a temperatura.")

  print("-" * 40)

  if not temperaturas:
    print("Nenhuma temperatura registrada.")
    return

  temperatura_media = sum(temperaturas) / len(temperaturas)
  temperatura_mais_alta = max(temperaturas)
  temperatura_mais_baixa = min(temperaturas)

  print(f"Temperatura média do dia: {temperatura_media:.2f}°C")
  print(f"Temperatura mais alta registrada: {temperatura_mais_alta}°C")
  print(f"Temperatura mais baixa registrada: {temperatura_mais_baixa}°C")

  print("-" * 40)

  alertas = False
  for hora, temp in enumerate(temperaturas):
    if temp > temperatura_maxima_alerta:
      print(f"ALERTA: Temperatura na hora {hora + 1} ({temp}°C) ultrapassou {temperatura_maxima_alerta}°C!")
      alertas = True
    elif temp < temperatura_minima_alerta:
      print(f"ALERTA: Temperatura na hora {hora + 1} ({temp}°C) ficou abaixo de {temperatura_minima_alerta}°C!")
      alertas = True

  if not alertas:
    print("Nenhum alerta de temperatura registrado.")

monitorar_temperatura()
```

```python
#8 - Desenvolva um algoritmo que auxilie na organização de uma biblioteca pessoal. O programa deve permitir que o usuário insira o título e o autor de diversos livros. Após a inserção, o sistema deve oferecer a opção de buscar por livros de um autor específico e listar todos os títulos correspondentes. O processo deve continuar até que o usuário decida encerrar.

def organizar_biblioteca():
  biblioteca = {}

  print("Sistema de Organização de Biblioteca Pessoal")

  while True:
    print("Escolha uma opção:")
    print("1 - Adicionar livro")
    print("2 - Buscar livros por autor")
    print("3 - Sair")

    opcao = input("Digite o número da opção desejada: ")

    if opcao == '1':
      titulo = input("Digite o título do livro: ")
      autor = input("Digite o nome do autor: ")
      biblioteca[titulo] = autor
      print(f"Livro '{titulo}' de '{autor}' adicionado à biblioteca.")
    elif opcao == '2':
      autor_busca = input("Digite o nome do autor para buscar: ")
      livros_do_autor = [titulo for titulo, autor in biblioteca.items() if autor.lower() == autor_busca.lower()]

      if livros_do_autor:
        print(f"Livros de '{autor_busca}':")
        for titulo in livros_do_autor:
          print(f"- {titulo}")
      else:
        print(f"Nenhum livro encontrado de '{autor_busca}'.")
    elif opcao == '3':
      print("Encerrando o sistema da biblioteca. Até mais!")
      break
    else:
      print("Opção inválida. Por favor, escolha uma opção válida.")

organizar_biblioteca()
```

```python
#9 - Elabore um algoritmo que simule um sistema de gerenciamento de tarefas domésticas. O programa deve permitir que o usuário insira diferentes tarefas e atribua um nível de prioridade a cada uma (alto, médio ou baixo). O sistema deve então exibir as tarefas organizadas por ordem de prioridade, começando pelas de alta prioridade, e permitir que o usuário marque tarefas como concluídas.

def gerenciar_tarefas():
    tarefas = []

    while True:
        print("Sistema de Gerenciamento de Tarefas Domésticas")
        print("1 - Adicionar tarefa")
        print("2 - Listar tarefas por prioridade")
        print("3 - Marcar tarefa como concluída")
        print("4 - Sair")

        opcao = input("Escolha uma opção: ")

        if opcao == '1':
            descricao = input("Digite a descrição da tarefa: ")
            while True:
                prioridade = input("Digite a prioridade (alta, media, baixa): ").lower()
                if prioridade in ['alta', 'media', 'baixa']:
                    break
                else:
                    print("Prioridade inválida. Use 'alta', 'media' ou 'baixa'.")
            tarefas.append({'descricao': descricao, 'prioridade': prioridade, 'concluida': False})
            print("Tarefa adicionada.")

        elif opcao == '2':
            if not tarefas:
                print("Nenhuma tarefa adicionada ainda.")
                continue

            prioridade_ordem = {'alta': 0, 'media': 1, 'baixa': 2}
            tarefas_ordenadas = sorted(tarefas, key=lambda x: prioridade_ordem[x['prioridade']])

            print("Lista de Tarefas (por Prioridade):")
            for i, tarefa in enumerate(tarefas_ordenadas):
                status = "Concluída" if tarefa['concluida'] else "Pendente"
                print(f"{i + 1}. [Prioridade: {tarefa['prioridade'].capitalize()}] {tarefa['descricao']} - Status: {status}")

        elif opcao == '3':
            if not tarefas:
                print("Nenhuma tarefa para marcar como concluída.")
                continue

            print("Tarefas:")
            for i, tarefa in enumerate(tarefas):
                status = "Concluída" if tarefa['concluida'] else "Pendente"
                print(f"{i + 1}. {tarefa['descricao']} - Status: {status}")

            try:
                indice_concluir = int(input("Digite o número da tarefa a ser marcada como concluída: ")) - 1
                if 0 <= indice_concluir < len(tarefas):
                    if tarefas[indice_concluir]['concluida']:
                        print("Esta tarefa já está marcada como concluída.")
                    else:
                        tarefas[indice_concluir]['concluida'] = True
                        print("Tarefa marcada como concluída.")
                else:
                    print("Número de tarefa inválido.")
            except ValueError:
                print("Entrada inválida. Por favor, digite um número.")

        elif opcao == '4':
            print("Encerrando o sistema de gerenciamento de tarefas. Até mais!")
            break

        else:
            print("Opção inválida. Por favor, escolha uma opção válida.")

gerenciar_tarefas()
```

```python
#10 - Crie um algoritmo que simule um sistema de acompanhamento de hábitos saudáveis. O programa deve permitir que o usuário registre diariamente a quantidade de copos de água ingeridos e o número de minutos dedicados a atividades físicas. Ao final de uma semana, o sistema deve calcular a média diária de consumo de água e de exercícios, compará-las com as recomendações de saúde (por exemplo, 8 copos de água e 30 minutos de exercício por dia), e fornecer um feedback sobre o desempenho do usuário.

def acompanhar_habitos():
    dias_semana = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"]
    consumo_agua = []
    minutos_exercicio = []

    print("Sistema de Acompanhamento de Hábitos Saudáveis")
    print("Registre seus hábitos durante a semana.")
    print("-" * 40)

    for dia in dias_semana:
        while True:
            try:
                agua = int(input(f"Quantos copos de água você ingeriu no {dia}? "))
                if agua >= 0:
                    consumo_agua.append(agua)
                    break
                else:
                    print("Por favor, insira um número não negativo.")
            except ValueError:
                print("Entrada inválida. Por favor, digite um número inteiro.")

        while True:
            try:
                exercicio = int(input(f"Quantos minutos de atividade física você fez no {dia}? "))
                if exercicio >= 0:
                    minutos_exercicio.append(exercicio)
                    break
                else:
                    print("Por favor, insira um número não negativo.")
            except ValueError:
                print("Entrada inválida. Por favor, digite um número inteiro.")
        print("-" * 20)

    if not consumo_agua or not minutos_exercicio:
        print("Dados insuficientes para calcular as médias.")
        return

    media_agua = sum(consumo_agua) / len(consumo_agua)
    media_exercicio = sum(minutos_exercicio) / len(minutos_exercicio)

    recomendacao_agua = 8
    recomendacao_exercicio = 30

    print("Resumo da Semana:")
    print(f"Média diária de consumo de água: {media_agua:.2f} copos")
    print(f"Média diária de minutos de exercício: {media_exercicio:.2f} minutos")
    print("-" * 40)

    print("Feedback:")
    if media_agua >= recomendacao_agua:
        print(f"- Excelente! Você atingiu ou superou a recomendação de {recomendacao_agua} copos de água por dia.")
    else:
        print(f"- Considere aumentar seu consumo de água. A recomendação é de {recomendacao_agua} copos por dia.")

    if media_exercicio >= recomendacao_exercicio:
        print(f"- Ótimo! Você atingiu ou superou a recomendação de {recomendacao_exercicio} minutos de exercício por dia.")
    else:
        print(f"- Tente dedicar mais tempo à atividade física. A recomendação é de {recomendacao_exercicio} minutos por dia.")

    print("Continue acompanhando seus hábitos para uma vida mais saudável!")

acompanhar_habitos()
```
