```python
#1 Crie um algoritmo que receba o horário (hora e minuto) de entrada e saída de um funcionário em um dia de trabalho. Calcule o tempo total trabalhado em horas e minutos, considerando que ele teve uma pausa de 1h para almoço, e diga se ele cumpriu uma jornada de pelo menos 8h diárias.

def calcular_tempo_trabalhado(hora_entrada, minuto_entrada, hora_saida, minuto_saida):

    # Converter tudo para minutos
    entrada_total_min = hora_entrada * 60 + minuto_entrada
    saida_total_min = hora_saida * 60 + minuto_saida

    # Calcular tempo total de trabalho em minutos
    tempo_total_min = saida_total_min - entrada_total_min

    # Subtrair 1 hora (60 minutos) de pausa para almoço
    tempo_trabalho_min = tempo_total_min - 60

    # Converter de volta para horas e minutos
    horas_trabalho = tempo_trabalho_min // 60
    minutos_trabalho = tempo_trabalho_min % 60

    # Verificar se cumpriu 8h de trabalho (480 minutos)
    cumpriu_jornada = tempo_trabalho_min >= 480

    # Exibir resultados
    print(f"Tempo trabalhado: {horas_trabalho}h {minutos_trabalho}min")
    if cumpriu_jornada:
        print("Jornada de trabalho cumprida (mínimo de 8h).")
    else:
        print("Jornada de trabalho NÃO cumprida.")

calcular_tempo_trabalhado(8, 30, 18, 0)
```

```python
#2 Crie um algoritmo que receba a data de nascimento (dia, mês e ano) de uma pessoa e a data atual, e diga com precisão sua idade em anos, meses e dias. Use condicionais para lidar com casos em que o mês ou dia ainda não tenha sido atingido no ano corrente.

def calcular_idade(dia_nasc, mes_nasc, ano_nasc, dia_atual, mes_atual, ano_atual):
    # Inicialmente calculamos a diferença bruta
    anos = ano_atual - ano_nasc
    meses = mes_atual - mes_nasc
    dias = dia_atual - dia_nasc

    # Ajustar dias e meses se necessário
    if dias < 0:
        meses -= 1
        # Para simplificar, vamos assumir que o mês anterior tem 30 dias
        dias += 30

    if meses < 0:
        anos -= 1
        meses += 12

    # Exibir o resultado
    print(f"Idade: {anos} anos, {meses} meses e {dias} dias.")

calcular_idade(15, 9, 1998, 26, 4, 2025)
```

```python
#3 Elabore um algoritmo que simule um monitor de qualidade do ar. O usuário informará 3 valores de índice de poluentes de diferentes horários de um mesmo dia. O programa deve calcular a média e classificar como "bom", "moderado", "ruim", "muito ruim" ou "péssimo", com base em faixas fornecidas pelo código.

def monitor_qualidade_ar():
    """
    Simula um monitor de qualidade do ar, calculando a média de três índices de poluentes
    e classificando a qualidade do ar.
    """

    indices = []
    for i in range(3):
        while True:
            try:
                indice = float(input(f"Digite o índice de poluentes do horário {i+1}: "))
                if indice >= 0:  # Índice de poluição não pode ser negativo
                    indices.append(indice)
                    break
                else:
                    print("O índice de poluentes não pode ser negativo. Tente novamente.")
            except ValueError:
                print("Entrada inválida. Digite um número.")

    media = sum(indices) / len(indices)

    if media <= 50:
        classificacao = "bom"
    elif media <= 100:
        classificacao = "moderado"
    elif media <= 150:
        classificacao = "ruim"
    elif media <= 200:
        classificacao = "muito ruim"
    else:
        classificacao = "péssimo"

    print(f"Índice de poluentes médio: {media:.2f}")
    print(f"Classificação da qualidade do ar: {classificacao}")


monitor_qualidade_ar()
```

```python
#4 Desenvolva um algoritmo que simule a geração de um extrato de abastecimento de combustível. O usuário deve informar o valor em reais que abasteceu, o preço do litro e a quilometragem inicial e final da viagem. Calcule quantos litros foram colocados e a média de km/l, informando se o carro teve desempenho bom (acima de 12 km/l), médio (8-12 km/l) ou ruim (abaixo de 8 km/l).

def calcular_extrato_abastecimento():
    """
    Simula a geração de um extrato de abastecimento de combustível.
    """

    try:
        valor_abastecido = float(input("Digite o valor em reais abastecido: "))
        preco_litro = float(input("Digite o preço do litro do combustível: "))
        km_inicial = float(input("Digite a quilometragem inicial da viagem: "))
        km_final = float(input("Digite a quilometragem final da viagem: "))

        if valor_abastecido <= 0 or preco_litro <= 0 or km_inicial < 0 or km_final < 0 or km_inicial >= km_final:
            print("Valores inválidos. Certifique-se de que os valores inseridos sejam positivos e que a quilometragem final seja maior que a inicial.")
            return

        litros_abastecidos = valor_abastecido / preco_litro
        distancia_percorrida = km_final - km_inicial
        consumo_medio = distancia_percorrida / litros_abastecidos


        print(f"--- Extrato de Abastecimento ---")
        print(f"Valor abastecido: R$ {valor_abastecido:.2f}")
        print(f"Preço do litro: R$ {preco_litro:.2f}")
        print(f"Litros abastecidos: {litros_abastecidos:.2f} L")
        print(f"Quilometragem inicial: {km_inicial:.2f} km")
        print(f"Quilometragem final: {km_final:.2f} km")
        print(f"Distância percorrida: {distancia_percorrida:.2f} km")
        print(f"Consumo médio: {consumo_medio:.2f} km/l")

        if consumo_medio > 12:
            print("Desempenho do carro: Bom")
        elif consumo_medio >= 8:
            print("Desempenho do carro: Médio")
        else:
            print("Desempenho do carro: Ruim")

    except ValueError:
        print("Entrada inválida. Certifique-se de inserir valores numéricos.")


calcular_extrato_abastecimento()
```

```python
#5 Crie um algoritmo que pergunte o nome e o tempo total (em minutos) gasto por uma pessoa em atividades físicas durante 7 dias. Ao final, mostre a média diária, a soma total e diga se ela atingiu o mínimo recomendado de 150 minutos semanais.

def atividade_fisica():
    """
    Calcula a média diária, a soma total de minutos de atividades físicas em 7 dias,
    e verifica se a meta mínima semanal foi atingida.
    """
    total_minutos = 0
    for dia in range(1, 8):
        while True:
            try:
                minutos = int(input(f"Digite o tempo gasto em atividades físicas no dia {dia} (em minutos): "))
                if minutos >= 0:
                    total_minutos += minutos
                    break
                else:
                    print("O tempo gasto não pode ser negativo. Tente novamente.")
            except ValueError:
                print("Entrada inválida. Digite um número inteiro.")
    media_diaria = total_minutos / 7
    print(f"Tempo total gasto em atividades físicas: {total_minutos} minutos")
    print(f"Média diária de atividades físicas: {media_diaria:.2f} minutos")
    if total_minutos >= 150:
        print("Você atingiu o mínimo recomendado de 150 minutos semanais de atividades físicas!")
    else:
        print("Você NÃO atingiu o mínimo recomendado de 150 minutos semanais de atividades físicas.")

atividade_fisica()
```

```python
#6 Desenvolva um algoritmo que simule o acompanhamento de consumo de água em uma residência. O usuário informará o valor do hidrômetro no início e no fim do mês. O sistema calculará o total consumido e o valor a pagar, seguindo uma tabela de faixas de consumo (ex: até 10 m³, tarifa fixa; de 11 a 20, valor extra por m³; acima de 20, tarifa mais alta).

def calcular_consumo_agua(inicio, fim):
    consumo = fim - inicio
    valor_pagar = 0

    if consumo <= 10:
        valor_pagar = 20  # Tarifa fixa para até 10m³
    elif consumo <= 20:
        valor_pagar = 20 + (consumo - 10) * 2  # Valor extra por m³ (de 11 a 20m³)
    else:
        valor_pagar = 20 + 10 * 2 + (consumo - 20) * 5  # Tarifa mais alta acima de 20m³

    print(f"Consumo total: {consumo} m³")
    print(f"Valor a pagar: R$ {valor_pagar:.2f}")

inicio = int(input("Digite o valor do hidrômetro no início do mês: "))
fim = int(input("Digite o valor do hidrômetro no fim do mês: "))

calcular_consumo_agua(inicio, fim)
```

```python
#7 Elabore um algoritmo que simule a pontualidade de voos. Receba o horário previsto e o horário real de chegada (hora e minuto). Informe o tempo de atraso ou adiantamento e classifique o voo como “pontual” (diferença ≤ 5 min), “ligeiramente atrasado/adiantado” (até 30 min) ou “muito atrasado/adiantado” (> 30 min).

def analisar_pontualidade_voo(hora_prevista, minuto_previsto, hora_real, minuto_real):
    """
    Analisa a pontualidade de um voo com base nos horários previstos e reais.
    """

    # Converter tudo para minutos
    previsto_total_min = hora_prevista * 60 + minuto_previsto
    real_total_min = hora_real * 60 + minuto_real

    # Calcular a diferença em minutos
    diferenca_minutos = real_total_min - previsto_total_min

    # Determinar o tempo de atraso ou adiantamento
    if diferenca_minutos == 0:
        tempo = "no horário"
    elif diferenca_minutos > 0:
        tempo = f"{diferenca_minutos} minutos de atraso"
    else:
        tempo = f"{abs(diferenca_minutos)} minutos de adiantamento"

    # Classificar a pontualidade do voo
    if abs(diferenca_minutos) <= 5:
        classificacao = "pontual"
    elif abs(diferenca_minutos) <= 30:
        classificacao = "ligeiramente atrasado/adiantado"
    else:
        classificacao = "muito atrasado/adiantado"

    # Exibir os resultados
    print(f"Horário previsto: {hora_prevista:02d}:{minuto_previsto:02d}")
    print(f"Horário real: {hora_real:02d}:{minuto_real:02d}")
    print(f"Tempo de diferença: {tempo}")
    print(f"Classificação do voo: {classificacao}")

analisar_pontualidade_voo(14, 0, 14, 10)  # ligeiramente atrasado
```

```python
#8 Crie um algoritmo que simule o cálculo de tempo total de tela (TV, celular, computador). Para cada dispositivo, o usuário informa quanto tempo usou por dia (em horas e minutos) por 3 dias. O sistema deve apresentar o tempo total em horas e minutos, destacando o aparelho com maior uso e emitindo um alerta caso o total passe de 10 horas diárias.

def calcular_tempo_tela():
    """
    Calcula o tempo total de tela para TV, celular e computador ao longo de 3 dias.
    """

    tempos = {"TV": [], "Celular": [], "Computador": []}
    dispositivos = list(tempos.keys())

    for dispositivo in dispositivos:
        for dia in range(1, 4):
            while True:
                try:
                    horas = int(input(f"Digite as horas de uso de {dispositivo} no dia {dia}: "))
                    minutos = int(input(f"Digite os minutos de uso de {dispositivo} no dia {dia}: "))
                    if 0 <= horas <= 24 and 0 <= minutos <= 59:
                        tempos[dispositivo].append(horas * 60 + minutos)
                        break
                    else:
                        print("Valores inválidos. As horas devem estar entre 0 e 24 e os minutos entre 0 e 59.")
                except ValueError:
                    print("Entrada inválida. Digite números inteiros para horas e minutos.")

    totais = {}
    for dispositivo in dispositivos:
      totais[dispositivo] = sum(tempos[dispositivo])

    tempo_total_minutos = sum(totais.values())
    tempo_total_horas = tempo_total_minutos // 60
    tempo_total_minutos = tempo_total_minutos % 60

    maior_uso = max(totais, key=totais.get)  # Encontra o dispositivo com maior uso

    print(f"Tempo total de tela: {tempo_total_horas} horas e {tempo_total_minutos} minutos.")
    print(f"Maior tempo de uso: {maior_uso} ({totais[maior_uso] // 60} horas e {totais[maior_uso] % 60} minutos)")

    if tempo_total_horas > 10 :
        print("ALERTA: Tempo total de tela excedeu 10 horas diárias.")

calcular_tempo_tela()
```

```python
#10 Crie um algoritmo que simule um plano de leitura. O usuário informa a quantidade de páginas de um livro, o tempo diário disponível para leitura (em minutos) e a velocidade média de leitura (páginas por minuto). O programa deve estimar quantos dias ele levará para terminar e sugerir um cronograma diário.

import math #para arredondar o número de dias para cima — isso evita "meio-dia" de leitura.

def calcular_plano_leitura():
    """
    Simula um plano de leitura, estimando o tempo necessário para terminar um livro
    e sugerindo um cronograma diário.
    """
    try:
        paginas_totais = int(input("Digite a quantidade de páginas do livro: "))
        tempo_diario_minutos = int(input("Digite o tempo diário disponível para leitura (em minutos): "))
        velocidade_leitura = float(input("Digite a velocidade média de leitura (páginas por minuto): "))

        if paginas_totais <= 0 or tempo_diario_minutos <= 0 or velocidade_leitura <= 0:
            print("Valores inválidos. Certifique-se de que todos os valores sejam positivos.")
            return

        paginas_por_dia = tempo_diario_minutos * velocidade_leitura
        dias_estimados = math.ceil(paginas_totais / paginas_por_dia)

        print("--- Plano de Leitura ---")
        print(f"Tempo estimado para terminar o livro: {dias_estimados} dias")

        print("Cronograma diário sugerido:")
        paginas_restantes = paginas_totais

        for dia in range(1, dias_estimados + 1):
            paginas_ler_hoje = min(paginas_por_dia, paginas_restantes)
            print(f"Dia {dia}: Ler {paginas_ler_hoje:.0f} páginas")
            paginas_restantes -= paginas_ler_hoje

    except ValueError:
        print("Entrada inválida. Certifique-se de digitar valores numéricos.")

calcular_plano_leitura()
```

```python
#12 Elabore um algoritmo que ajude no controle de validade de alimentos. O usuário informa a data de fabricação e a validade (em dias). O programa calcula quantos dias faltam para vencer e emite alertas em cores (texto) se estiver perto de vencer (menos de 5 dias) ou já vencido.

from datetime import datetime, timedelta

# Entrada de dados
fabricacao_str = input("Informe a data de fabricação (dd/mm/aaaa): ")
validade_dias = int(input("Informe a validade (em dias): "))

# Conversão da data de fabricação
fabricacao = datetime.strptime(fabricacao_str, "%d/%m/%Y")

# Cálculo da data de vencimento
vencimento = fabricacao + timedelta(days=validade_dias)

# Data de hoje
hoje = datetime.now()

# Cálculo dos dias restantes
dias_restantes = (vencimento - hoje).days

# Exibição com alertas de cor
if dias_restantes < 0:
    print("\033[91mProduto VENCIDO há", abs(dias_restantes), "dias!\033[0m")
elif dias_restantes < 5:
    print("\033[93mAtenção: Faltam", dias_restantes, "dias para vencer.\033[0m")
else:
    print("\033[92mProduto válido. Faltam", dias_restantes, "dias para vencer.\033[0m")
```

```python
#13 Crie um algoritmo que receba a data atual e a data de um evento (em qualquer ano) e informe quantos dias faltam. O programa deve considerar se o evento está no mesmo ano, em um ano futuro ou se já passou.

from datetime import datetime, timedelta

def dias_para_evento(data_atual_str, data_evento_str):
    """
    Calcula a diferença em dias entre a data atual e a data de um evento.

    Args:
        data_atual_str: String representando a data atual (formato DD/MM/AAAA).
        data_evento_str: String representando a data do evento (formato DD/MM/AAAA).

    Returns:
        Uma string informando quantos dias faltam para o evento, considerando
        se o evento já passou, está no mesmo ano ou em um ano futuro.
    """
    try:
        data_atual = datetime.strptime(data_atual_str, "%d/%m/%Y")
        data_evento = datetime.strptime(data_evento_str, "%d/%m/%Y")

        diferenca = data_evento - data_atual
        dias_restantes = diferenca.days

        if dias_restantes < 0:
            return f"O evento já passou. Se passaram {abs(dias_restantes)} dias."
        elif dias_restantes == 0:
            return "O evento é hoje!"
        else:
            return f"Faltam {dias_restantes} dias para o evento."

    except ValueError:
        return "Formato de data inválido. Use o formato DD/MM/AAAA."

data_atual = input("Digite a data atual (DD/MM/AAAA): ")
data_evento = input("Digite a data do evento (DD/MM/AAAA): ")

resultado = dias_para_evento(data_atual, data_evento)
resultado
```

```python
#15 Elabore um algoritmo que calcule o tempo exato de um voo com escalas. O usuário informa o horário de saída, o tempo de voo e o tempo total de espera em escalas. O sistema deve calcular o horário final de chegada e a duração total da viagem.

from datetime import datetime, timedelta

def calcular_tempo_voo():
    """
    Calcula o tempo exato de um voo com escalas.
    """
    try:
        # Entrada de dados
        saida_str = input("Informe o horário de saída (formato HH:MM): ")
        tempo_voo_horas = int(input("Informe o tempo de voo (em horas): "))
        tempo_voo_minutos = int(input("Informe o tempo de voo (em minutos): "))
        tempo_espera_total_minutos = int(input("Informe o tempo total de espera em escalas (em minutos): "))

        # Conversão do horário de saída para um objeto datetime
        saida = datetime.strptime(saida_str, "%H:%M")

        # Cálculo do tempo total de voo em minutos
        tempo_voo_total_minutos = tempo_voo_horas * 60 + tempo_voo_minutos

        # Cálculo do tempo total da viagem em minutos
        tempo_total_minutos = tempo_voo_total_minutos + tempo_espera_total_minutos

        # Cálculo da hora de chegada
        chegada = saida + timedelta(minutes=tempo_total_minutos)

        # Formatação da saída
        duracao_horas = tempo_total_minutos // 60
        duracao_minutos = tempo_total_minutos % 60

        print(f"Horário de chegada: {chegada.strftime('%H:%M')}")
        print(f"Duração total da viagem: {duracao_horas} horas e {duracao_minutos} minutos")

    except ValueError:
        print("Formato de entrada inválido. Certifique-se de seguir as instruções.")

calcular_tempo_voo()
```

```python
#21 Crie um algoritmo que simule o tempo de preparo de uma refeição. O usuário escolhe 3 pratos (entrada, principal, sobremesa), cada um com seu tempo de preparo. O sistema informa o tempo total e sugere uma ordem de preparo para otimizar o tempo, se possível.

def calcular_tempo_refeicao():
    """
    Simula o tempo de preparo de uma refeição, considerando três pratos.
    """
    pratos = {
        "entrada": 0,
        "principal": 0,
        "sobremesa": 0
    }
    for prato in pratos:
        while True:
            try:
                tempo = int(input(f"Digite o tempo de preparo do(a) {prato} (em minutos): "))
                if tempo > 0:
                    pratos[prato] = tempo
                    break
                else:
                    print("Tempo de preparo inválido. Deve ser um valor positivo.")
            except ValueError:
                print("Entrada inválida. Digite um número inteiro.")

    tempo_total = sum(pratos.values())
    print(f"Tempo total de preparo da refeição: {tempo_total} minutos")

    # Sugestão de ordem de preparo (otimização)
    ordem = sorted(pratos, key=pratos.get) #ordena pelo tempo de preparo

    print("Sugestão de ordem de preparo (para otimizar tempo):")
    for prato in ordem:
        print(f"- {prato} ({pratos[prato]} minutos)")

calcular_tempo_refeicao()
```

```python
#22 Elabore um algoritmo que ajude um motorista a planejar uma viagem de ida e volta. Receba a distância (em km), o tempo de parada previsto na cidade de destino e a velocidade média. Calcule o tempo total da jornada e o tempo disponível para descanso antes do retorno.

def calcular_viagem(distancia_km, tempo_parada_minutos, velocidade_media_kmh):
    """
    Calcula o tempo total de uma viagem de ida e volta, considerando o tempo de parada.

    Args:
        distancia_km: A distância de ida em km.
        tempo_parada_minutos: O tempo de parada na cidade de destino em minutos.
        velocidade_media_kmh: A velocidade média em km/h.
    """

    try:
        # Validações
        if distancia_km <= 0 or tempo_parada_minutos < 0 or velocidade_media_kmh <= 0:
            raise ValueError("Valores inválidos. Certifique-se de inserir valores positivos para distância, tempo de parada e velocidade.")

        # Tempo de viagem de ida
        tempo_ida_horas = distancia_km / velocidade_media_kmh
        tempo_ida_minutos = tempo_ida_horas * 60


        # Tempo total da viagem de ida e volta (sem parada)
        tempo_total_ida_volta = tempo_ida_minutos * 2

        # Tempo total da jornada (ida, volta e parada)
        tempo_total_minutos = tempo_total_ida_volta + tempo_parada_minutos
        tempo_total_horas = tempo_total_minutos // 60
        tempo_total_minutos_restantes = tempo_total_minutos % 60

        print(f"Tempo total da jornada (ida e volta com parada): {tempo_total_horas} horas e {tempo_total_minutos_restantes} minutos.")

        # Tempo disponível para descanso antes do retorno
        tempo_descanso_minutos = tempo_parada_minutos
        tempo_descanso_horas = tempo_descanso_minutos // 60
        tempo_descanso_minutos_restantes = tempo_descanso_minutos % 60
        print(f"Tempo disponível para descanso antes do retorno: {tempo_descanso_horas} horas e {tempo_descanso_minutos_restantes} minutos.")

    except ValueError as e:
        print(f"Erro: {e}")


# Solicita informações ao usuário
distancia = float(input("Digite a distância (em km): "))
tempo_parada = int(input("Digite o tempo de parada previsto (em minutos): "))
velocidade_media = float(input("Digite a velocidade média (em km/h): "))

calcular_viagem(distancia, tempo_parada, velocidade_media)
```

```python
#23 Desenvolva um algoritmo que calcule quantos segundos faltam para o ano novo, recebendo a data e hora atual. O sistema deve considerar se o ano é bissexto ou não.

from datetime import datetime

def segundos_para_ano_novo():
    try:
        # Entrada de dados
        data_atual_str = input("Informe a data e hora atual (formato dd/mm/aaaa HH:MM:SS): ")
        data_atual = datetime.strptime(data_atual_str, "%d/%m/%Y %H:%M:%S")

        # Ano atual
        ano_atual = data_atual.year

        # Verificar se o ano é bissexto
        bissexto = False
        if (ano_atual % 4 == 0 and ano_atual % 100 != 0) or (ano_atual % 400 == 0):
            bissexto = True

        # Definir a data do próximo ano novo
        ano_novo = datetime(year=ano_atual + 1, month=1, day=1, hour=0, minute=0, second=0)

        # Calcular a diferença em segundos
        diferenca = ano_novo - data_atual
        segundos_restantes = int(diferenca.total_seconds())

        print(f"Faltam {segundos_restantes} segundos para o Ano Novo!")
        print(f"O ano atual {'é' if bissexto else 'não é'} bissexto.")

    except ValueError:
        print("Formato de entrada inválido. Use o formato dd/mm/aaaa HH:MM:SS.")

segundos_para_ano_novo()
```

```python
#24 Crie um algoritmo que simule a leitura de um termômetro inteligente. O usuário informa temperaturas de manhã, tarde e noite por 5 dias. O sistema deve exibir a maior e menor temperatura da semana, a média geral e se houve risco de insolação (temperaturas acima de 35ºC em dois períodos do mesmo dia).

def monitor_temperatura():
    """
    Simula a leitura de um termômetro inteligente por 5 dias,
    coletando temperaturas de manhã, tarde e noite.
    """
    temperaturas = []
    for dia in range(5):
        print(f"Dia {dia + 1}:")
        for periodo in ["manhã", "tarde", "noite"]:
            while True:
                try:
                    temperatura = float(input(f"Digite a temperatura da {periodo}: "))
                    temperaturas.append(temperatura)
                    break
                except ValueError:
                    print("Entrada inválida. Digite um número válido.")

    maior_temperatura = max(temperaturas)
    menor_temperatura = min(temperaturas)
    media_geral = sum(temperaturas) / len(temperaturas)

    risco_insolacao = False
    for dia in range(5):
      if temperaturas[dia*3] > 35 and temperaturas[dia*3+1] > 35:
        risco_insolacao = True
        break
      elif temperaturas[dia*3] > 35 and temperaturas[dia*3+2] > 35:
        risco_insolacao = True
        break
      elif temperaturas[dia*3+1] > 35 and temperaturas[dia*3+2] > 35:
        risco_insolacao = True
        break

    print("Resultados:")
    print(f"Maior temperatura da semana: {maior_temperatura:.1f}°C")
    print(f"Menor temperatura da semana: {menor_temperatura:.1f}°C")
    print(f"Média geral das temperaturas: {media_geral:.1f}°C")
    if risco_insolacao:
        print("Houve risco de insolação.")
    else:
        print("Não houve risco de insolação.")

monitor_temperatura()
```

```python
#27 Desenvolva um algoritmo que receba a hora e minuto em que um paciente tomou um remédio e calcule os próximos 3 horários ideais para as próximas doses, considerando que o medicamento deve ser tomado a cada 8 horas.

def calcular_proximos_horarios(hora_inicial, minuto_inicial):
    """
    Calcula os próximos 3 horários ideais para as próximas doses de um medicamento,
    considerando um intervalo de 8 horas entre as doses.

    Args:
        hora_inicial: A hora em que o medicamento foi tomado inicialmente.
        minuto_inicial: O minuto em que o medicamento foi tomado inicialmente.
    """

    try:
        hora_inicial = int(hora_inicial)
        minuto_inicial = int(minuto_inicial)

        if not (0 <= hora_inicial <= 23 and 0 <= minuto_inicial <= 59):
            raise ValueError("Horário inválido. Digite um horário válido (HH:MM)")

        horario_inicial = datetime.now().replace(hour=hora_inicial, minute=minuto_inicial, second=0, microsecond=0)
        proximos_horarios = []

        for i in range(1, 4):
            proximo_horario = horario_inicial + timedelta(hours=8 * i)
            proximos_horarios.append(proximo_horario.strftime("%H:%M"))

        print("Próximos 3 horários para as doses:")
        for horario in proximos_horarios:
            print(horario)

    except ValueError as e:
        print(f"Erro: {e}")

# Solicita a hora e minuto da primeira dose ao usuário
hora_inicial = input("Digite a hora em que o medicamento foi tomado (HH): ")
minuto_inicial = input("Digite o minuto em que o medicamento foi tomado (MM): ")

calcular_proximos_horarios(hora_inicial, minuto_inicial)
```

```python
#29 Crie um algoritmo que ajude a calcular o tempo ideal de estudo. O usuário informa o total de horas disponíveis na semana e a quantidade de disciplinas. O sistema calcula quantas horas deve dedicar a cada matéria e alerta se alguma ficar com menos de 1h.

import math

def calcular_tempo_estudo(horas_disponiveis, num_disciplinas):
    """
    Calcula o tempo ideal de estudo por disciplina e emite alertas.
    """
    try:
        horas_por_disciplina = horas_disponiveis / num_disciplinas
        print("Tempo ideal de estudo por disciplina:")
        for i in range(num_disciplinas):
            if horas_por_disciplina < 1:
                print(f"Disciplina {i+1}: {horas_por_disciplina:.2f} horas (ALERTA: Menos de 1 hora!)")
            else:
                print(f"Disciplina {i+1}: {horas_por_disciplina:.2f} horas")
    except ZeroDivisionError:
        print("Erro: O número de disciplinas não pode ser zero.")
    except ValueError:
        print("Erro: Digite valores numéricos válidos.")


horas_totais = float(input("Digite o total de horas disponíveis na semana: "))
numero_disciplinas = int(input("Digite a quantidade de disciplinas: "))

calcular_tempo_estudo(horas_totais, numero_disciplinas)
```

```python
#30 Desenvolva um algoritmo que simule o acompanhamento de produtividade de um trabalhador remoto. Receba o número de tarefas concluídas por dia durante 5 dias e calcule a média. Se a produtividade cair por dois dias consecutivos, o sistema emite um alerta.

def monitorar_produtividade():
    """
    Monitora a produtividade de um trabalhador remoto por 5 dias,
    calculando a média de tarefas concluídas e emitindo alertas.
    """
    tarefas_diarias = []
    for dia in range(5):
        while True:
            try:
                tarefas = int(input(f"Digite o número de tarefas concluídas no dia {dia + 1}: "))
                if tarefas >= 0:
                    tarefas_diarias.append(tarefas)
                    break
                else:
                    print("Número de tarefas inválido. Deve ser não negativo.")
            except ValueError:
                print("Entrada inválida. Digite um número inteiro.")

    media_tarefas = sum(tarefas_diarias) / len(tarefas_diarias)
    print(f"Média de tarefas concluídas por dia: {media_tarefas:.2f}")

    queda_consecutiva = 0
    for i in range(1, len(tarefas_diarias)):
        if tarefas_diarias[i] < tarefas_diarias[i - 1]:
            queda_consecutiva += 1
            if queda_consecutiva >= 2:
                print("ALERTA: Produtividade caiu por dois dias consecutivos!")
                break
        else:
            queda_consecutiva = 0

monitorar_produtividade()
```
