```python
#1 - Crie um algoritmo que simule um sistema de reservas de um restaurante. O programa deve solicitar o número de pessoas para a reserva e verificar se há disponibilidade, considerando que o restaurante possui apenas mesas de 4 lugares. Se o número de pessoas exceder a capacidade de uma mesa, o programa deve calcular quantas mesas serão necessárias e informar ao usuário.

def reservar_mesa():

  try:
    num_pessoas = int(input("Digite o número de pessoas para a reserva: "))
    if num_pessoas <= 0:
      print("Número de pessoas inválido.")
      return

    mesas_necessarias = (num_pessoas + 3) // 4  # Calcula o número de mesas necessárias, arredondando para cima

    if mesas_necessarias == 1:
      print("Reserva realizada com sucesso! Uma mesa para", num_pessoas, "pessoas.")
    else:
      print(f"Reserva realizada com sucesso! Você precisará de {mesas_necessarias} mesas para {num_pessoas} pessoas.")

  except ValueError:
    print("Entrada inválida. Por favor, insira um número inteiro.")

reservar_mesa()
```

```python
#2 - Desenvolva um algoritmo que auxilie na gestão de estoque de uma loja. O programa deve permitir que o usuário insira o nome e a quantidade de diferentes produtos. Se a quantidade de um produto for inferior a 5 unidades, o sistema deve alertar que o estoque está baixo e sugerir a reposição. O processo deve continuar até que o usuário decida encerrar.

def gerenciar_estoque():
  estoque = {}

  while True:
    nome_produto = input("Digite o nome do produto ou 'sair' para encerrar: ")
    if nome_produto.lower() == 'sair':
      break

    try:
      quantidade = int(input("Digite a quantidade em estoque: "))
      if quantidade < 0:
        print("Quantidade inválida. Digite um valor não negativo.")
        continue

      estoque[nome_produto] = quantidade

      if quantidade < 5:
        print(f"Alerta: Estoque baixo para {nome_produto}. Reposição necessária!")

    except ValueError:
      print("Quantidade inválida. Digite um número inteiro.")

  print("Estoque atual:")
  for produto, quantidade in estoque.items():
    print(f"{produto}: {quantidade} unidades")


gerenciar_estoque()
```

```python
#3 - Elabore um algoritmo que simule um quiz de perguntas e respostas. O programa deve apresentar uma série de perguntas ao usuário, oferecer opções de resposta e verificar se a resposta fornecida está correta. Ao final, o sistema deve exibir a pontuação total do usuário e fornecer um feedback baseado no desempenho.

def quiz():
    perguntas = {
        "Qual a capital do Brasil?": "Brasília",
        "Qual o maior planeta do sistema solar?": "Júpiter",
        "Quanto é 2 + 2?": "4"
    }

    pontuacao = 0

    for pergunta, resposta_correta in perguntas.items():
        resposta_usuario = input(f"{pergunta} ")
        if resposta_usuario.lower() == resposta_correta.lower():
            print("Correto!")
            pontuacao += 1
        else:
            print(f"Incorreto. A resposta correta é {resposta_correta}.")

    print(f"Sua pontuação final é: {pontuacao} de {len(perguntas)}")

    if pontuacao == len(perguntas):
        print("Parabéns, você acertou todas as perguntas!")
    elif pontuacao >= len(perguntas) / 2:
        print("Bom desempenho! Você acertou mais da metade das perguntas.")
    else:
        print("Que pena, tente novamente!")

quiz()
```

```python
#4 - Crie um algoritmo que simule um sistema de controle de acesso a um edifício. O programa deve solicitar ao usuário um código de acesso e verificar se ele é válido. Caso contrário, deve permitir novas tentativas até que o número máximo de 3 tentativas seja atingido. Se o acesso for concedido, uma mensagem de boas-vindas deve ser exibida; caso contrário, uma mensagem de bloqueio deve ser apresentada.

def controlar_acesso():
    codigo_correto = "1234"  # Código de acesso correto
    tentativas = 0
    max_tentativas = 3

    while tentativas < max_tentativas:
        codigo_usuario = input("Digite o código de acesso: ")
        if codigo_usuario == codigo_correto:
            print("Acesso concedido! Bem-vindo(a)!")
            return  # Sai da função após o acesso bem-sucedido ou com o acesso bloqueado

        tentativas += 1
        tentativas_restantes = max_tentativas - tentativas
        if tentativas_restantes > 0:
            print(f"Código incorreto. Você tem mais {tentativas_restantes} tentativas.")
        else:
            print("Número máximo de tentativas atingido. Acesso bloqueado, tente novamente em 10 minutos")

controlar_acesso()
```

```python
#5 - Desenvolva um algoritmo que ajude no planejamento de uma viagem de carro. O programa deve solicitar a distância total a ser percorrida, a capacidade do tanque de combustível do veículo e o consumo médio de combustível (km/l). Com base nessas informações, o sistema deve calcular quantas paradas para reabastecimento serão necessárias, assumindo que o motorista sempre reabasteça o tanque completamente.

def calcular_paradas():
    try:
        distancia_total = float(input("Digite a distância total da viagem (em km): "))
        capacidade_tanque = float(input("Digite a capacidade do tanque de combustível (em litros): "))
        consumo_medio = float(input("Digite o consumo médio de combustível (em km/l): "))

        if distancia_total <= 0 or capacidade_tanque <= 0 or consumo_medio <= 0:
            print("Valores inválidos. Certifique-se de inserir valores positivos.")
            return

        autonomia = capacidade_tanque * consumo_medio
        numero_paradas = distancia_total // autonomia

        if distancia_total % autonomia != 0:
            numero_paradas += 1

        print(f"Serão necessárias {int(numero_paradas)} paradas para reabastecimento.")

    except ValueError:
        print("Entrada inválida. Por favor, insira números.")

calcular_paradas()
```

```python
#6 - Elabore um algoritmo que simule um sistema de avaliação de filmes. O programa deve permitir que o usuário insira o título de vários filmes e uma nota de 0 a 10 para cada um. Após o usuário finalizar as entradas, o sistema deve exibir a média das notas atribuídas e listar os filmes que receberam nota superior a 8, destacando-os como altamente recomendados.

def avaliar_filmes():
    filmes = {}
    while True:
        titulo = input("Digite o título do filme ou 'fim' para encerrar: ")
        if titulo.lower() == 'fim':
            break
        try:
            nota = float(input("Digite a nota (0-10): "))
            if 0 <= nota <= 10:
                filmes[titulo] = nota
            else:
                print("Nota inválida. Digite um valor entre 0 e 10.")
        except ValueError:
            print("Nota inválida. Digite um número.")


    if not filmes:
        print("Nenhum filme avaliado.")
        return

    soma_notas = sum(filmes.values())
    media = soma_notas / len(filmes)
    print(f"A média das notas dos filmes é: {media:.2f}")

    print("Filmes altamente recomendados (nota > 8):")
    for titulo, nota in filmes.items():
        if nota > 8:
            print(f"- {titulo} (Nota: {nota})")

avaliar_filmes()
```

```python
#7 Crie um algoritmo que simule um sistema de monitoramento de temperatura ambiente. O programa deve solicitar ao usuário que insira a temperatura registrada a cada hora durante um período de 24 horas. Ao final, o sistema deve calcular a temperatura média do dia, identificar a temperatura mais alta e a mais baixa registradas, e alertar se em algum momento a temperatura ultrapassou 30°C ou ficou abaixo de 15°C.

def monitoramento_temperatura():
    temperaturas = []

    print("Sistema de Monitoramento de Temperatura - 24 Horas")
    print("-" * 50)

    for hora in range(24):
        while True:
            try:
                temp = float(input(f"Digite a temperatura registrada às {hora:02d}:00: "))
                temperaturas.append(temp)
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número.")


    temp_media = sum(temperaturas) / len(temperaturas)
    temp_max = max(temperaturas)
    temp_min = min(temperaturas)


    alerta_alta = any(t > 30 for t in temperaturas)
    alerta_baixa = any(t < 15 for t in temperaturas)


    print("Relatório do Dia:")
    print(f"Temperatura média: {temp_media:.2f}°C")
    print(f"Temperatura mais alta: {temp_max:.2f}°C")
    print(f"Temperatura mais baixa: {temp_min:.2f}°C")

    if alerta_alta:
        print(" Alerta: Temperatura acima de 30°C registrada durante o dia.")
    if alerta_baixa:
        print(" Alerta: Temperatura abaixo de 15°C registrada durante o dia.")


monitoramento_temperatura()
```

```python
#8 Desenvolva um algoritmo que auxilie na organização de uma biblioteca pessoal. O programa deve permitir que o usuário insira o título e o autor de diversos livros. Após a inserção, o sistema deve oferecer a opção de buscar por livros de um autor específico e listar todos os títulos correspondentes. O processo deve continuar até que o usuário decida encerrar.

def organizar_biblioteca():
    biblioteca = {}

    while True:
        opcao = input("Digite 'a' para adicionar um livro, 'b' para buscar por autor, ou 's' para sair: ")

        if opcao.lower() == 'a':
            titulo = input("Digite o título do livro: ")
            autor = input("Digite o nome do autor: ")
            if autor in biblioteca:
                biblioteca[autor].append(titulo)
            else:
                biblioteca[autor] = [titulo]
            print(f"Livro '{titulo}' de {autor} adicionado com sucesso!")

        elif opcao.lower() == 'b':
            autor = input("Digite o nome do autor que deseja buscar: ")
            if autor in biblioteca:
                print(f"Livros de {autor}:")
                for titulo in biblioteca[autor]:
                    print(f"- {titulo}")
            else:
                print(f"Nenhum livro encontrado para o autor '{autor}'.")

        elif opcao.lower() == 's':
            print("Encerrando o programa...")
            break
        else:
            print("Opção inválida. Digite 'a', 'b' ou 's'.")

organizar_biblioteca()
```

```python
#9 Elabore um algoritmo que simule um sistema de gerenciamento de tarefas domésticas. O programa deve permitir que o usuário insira diferentes tarefas e atribua um nível de prioridade a cada uma (alto, médio ou baixo). O sistema deve então exibir as tarefas organizadas por ordem de prioridade, começando pelas de alta prioridade, e permitir que o usuário marque tarefas como concluídas.
tarefas = []

def exibir_menu():
    print("========= MENU =========")
    print("1. Adicionar tarefa")
    print("2. Listar tarefas por prioridade")
    print("3. Marcar tarefa como concluída")
    print("4. Sair")
    print("========================")

def adicionar_tarefa():
    nome = input("Digite a descrição da tarefa: ")
    prioridade = input("Digite a prioridade (alto, médio, baixo): ").lower()

    if prioridade not in ['alto', 'médio', 'baixo']:
        print("Prioridade inválida! Use: alto, médio ou baixo.")
        return

    tarefas.append({
        "nome": nome,
        "prioridade": prioridade,
        "concluida": False
    })
    print("Tarefa adicionada com sucesso!")

def listar_tarefas():
    print("=== Lista de Tarefas por Prioridade ===")
    prioridades = ['alto', 'médio', 'baixo']
    for nivel in prioridades:
        print(f"Prioridade {nivel.upper()}:")
        encontrou = False
        for i, tarefa in enumerate(tarefas):
            if tarefa["prioridade"] == nivel and not tarefa["concluida"]:
                print(f"{i + 1}. {tarefa['nome']}")
                encontrou = True
        if not encontrou:
            print("Nenhuma tarefa.")

def marcar_concluida():
    listar_tarefas()
    try:
        num = int(input("Digite o número da tarefa que deseja marcar como concluída: "))
        if 1 <= num <= len(tarefas):
            if tarefas[num - 1]["concluida"]:
                print("Essa tarefa já foi concluída.")
            else:
                tarefas[num - 1]["concluida"] = True
                print("Tarefa marcada como concluída!")
        else:
            print("Número inválido!")
    except ValueError:
        print("Entrada inválida! Digite um número.")

while True:
    exibir_menu()
    escolha = input("Escolha uma opção: ")

    if escolha == '1':
        adicionar_tarefa()
    elif escolha == '2':
        listar_tarefas()
    elif escolha == '3':
        marcar_concluida()
    elif escolha == '4':
        print("Saindo do sistema. Até mais!")
        break
    else:
        print("Opção inválida, tente novamente.")
```

```python
#10 Crie um algoritmo que simule um sistema de acompanhamento de hábitos saudáveis. O programa deve permitir que o usuário registre diariamente a quantidade de copos de água ingeridos e o número de minutos dedicados a atividades físicas. Ao final de uma semana, o sistema deve calcular a média diária de consumo de água e de exercícios, compará-las com as recomendações de saúde (por exemplo, 8 copos de água e 30 minutos de exercício por dia), e fornecer um feedback sobre o desempenho do usuário.

def acompanhar_habitos():
    """Simula um sistema de acompanhamento de hábitos saudáveis."""

    registro_diario = []

    for dia in range(7):
        while True:
            try:
                agua = int(input(f"Dia {dia + 1}: Quantos copos de água você bebeu hoje? "))
                exercicio = int(input(f"Dia {dia + 1}: Quantos minutos de exercício você fez hoje? "))
                if agua >= 0 and exercicio >= 0:
                    registro_diario.append({"agua": agua, "exercicio": exercicio})
                    break
                else:
                    print("Valores inválidos. Digite números não negativos.")
            except ValueError:
                print("Entrada inválida. Digite números inteiros.")


    agua_total = sum(dia["agua"] for dia in registro_diario)
    exercicio_total = sum(dia["exercicio"] for dia in registro_diario)

    media_agua = agua_total / 7
    media_exercicio = exercicio_total / 7


    print("--- Relatório Semanal ---")
    print(f"Média diária de água: {media_agua:.2f} copos")
    print(f"Média diária de exercícios: {media_exercicio:.2f} minutos")

    if media_agua >= 8:
        print("Parabéns! Você atingiu a meta de consumo de água diário.")
    else:
        print(f"Você está abaixo da meta de consumo de água (8 copos). Procure beber mais água!")

    if media_exercicio >= 30:
        print("Parabéns! Você atingiu a meta de exercícios diários.")
    else:
        print(f"Você está abaixo da meta de exercícios (30 minutos). Procure se exercitar mais!")

acompanhar_habitos()
```
